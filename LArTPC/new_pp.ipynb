```python
import sys
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
import os
import numpy as np
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
```

```python
# takes full raw data and extracts waveform of length nticks
def extract_wave(data, nticks=200):
    string = 'tck_'
    waveforms = []
    #Here I extract a column in each iteration and append to list
    for i in range(nticks):
        waveforms.append(data[string+str(i)].astype(np.int16))
    #convert to numpy ndarray
    waveforms = np.array(waveforms).astype(np.int16)
    #since raws and columns are inverted we need to transpose it
    return np.transpose(waveforms)

# takes full raw data and returns waveform of length nticks
# only keeps waves at a desired adc count 
def get_std_waveforms_per_adc(data_noisy, data_clean, nticks, desired_adc):
    #Extract and scale waveform data (passthrough rn)
    raw_waveforms_noisy = extract_wave(data_noisy, nticks)
    raw_waveforms_clean = extract_wave(data_clean, nticks)
    #print('before adc filter: ', raw_waveforms_noisy.shape, raw_waveforms_clean.shape)

    noisy_ = []
    clean_ = []

    for i, wave in enumerate(raw_waveforms_clean):
        if max(wave) == desired_adc:
            noisy_.append(raw_waveforms_noisy[i])
            clean_.append(wave)
    
    del raw_waveforms_noisy, raw_waveforms_clean

    noisy_ = np.array(noisy_)
    clean_ = np.array(clean_)

    #print('after adc filter: ', noisy_.shape, clean_.shape)
    #print(raw_waveforms) 
    #scaled_waveforms = waveform_scaler.fit_transform(raw_waveforms)
    return noisy_, clean_

# loads raw files, extracts waveforms and returns num_samples that have
# the desired_adc 
def get_waves_per_adc(desired_adc, noisy_filenames, clean_filenames, num_samples=100000, ticks=200):
    noisy_wave = np.empty((0,ticks))
    clean_wave = np.empty((0,ticks))

    for i in range(len(noisy_filenames)):
        noisy = np.load(noisy_filenames[i])
        clean = np.load(clean_filenames[i])

        std_noisy, std_clean = get_std_waveforms_per_adc(noisy, clean, ticks, desired_adc)
        del noisy, clean
        noisy_wave = np.concatenate((noisy_wave, std_noisy))
        clean_wave = np.concatenate((clean_wave, std_clean))

        del std_noisy, std_clean

        if noisy_wave.shape[0] > num_samples:
            break
    
    return noisy_wave, clean_wave
```

```python
wireplane = 'U'
nticks = 200
sig_name = "-signal"
cln_name = "-clnsig"
path = '/home/vlian/Workspace/ar39_samples/' + wireplane + '/'

ar_39_filenames = sorted([path+f for f in listdir(path) if (isfile(join(path, f)) and sig_name in f)])
ar_39_clean_filenames = sorted([path+f for f in listdir(path) if (isfile(join(path, f)) and cln_name in f)])

for i in range(10):
    ar_39_filenames, ar_39_clean_filenames = shuffle(ar_39_filenames, ar_39_clean_filenames)
print(len(ar_39_filenames), len(ar_39_clean_filenames))
```

```python
data_adc_5_15 = []
for i in range(11):
    noisy_WAVES, clean_WAVES = get_waves_per_adc(i+5, ar_39_filenames, ar_39_clean_filenames, num_samples=80000)
    data_adc_5_15.append([noisy_WAVES[0:80000], clean_WAVES[0:80000]])
    del noisy_WAVES, clean_WAVES
```

For each ADC, split into 50:50 train set, then add noise to each set

[ [adc5], [adc6], [adc7], [adc8],.. ] 

[adc5] -->  [ [x_train, y_train], [x_valid, y_valid], [x_test, y_test] ]

> Go through each adc group
> Load the same amount of noise samples
>>>> For each plane, there are 700 files
>>>> There are 1.4 million noise samples in total
>>>>>>> 2000 noise samples per file 

>>>>>>> Then for each adc group, we load 40 files at a time

```python
noise_path = '/home/vlian/Workspace/more-noise/' + wireplane + '/'
moise_name = '-noise'
noise_filenames = sorted([noise_path+f for f in listdir(noise_path) if (isfile(join(noise_path, f)) and moise_name in f)])
noise_filenames = shuffle(noise_filenames)
print(len(noise_filenames))
```

```python
dataset = []
noise_file_idx = 0
for adc_grp in data_adc_5_15:
    x_data = adc_grp[0].copy()
    y_data = adc_grp[1].copy()
    adc_grp[0], adc_grp[1] = [None]*2

    x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.4, shuffle=True)
    noise_samples = np.concatenate([np.load(fname) for fname in noise_filenames[noise_file_idx:noise_file_idx + 40]])
    noise_samples = extract_wave(noise_samples)
    noiseless_samples = noise_samples*0
    noise_file_idx += 40
    x_train_noise, x_test_noise, y_train_noise, y_test_noise = train_test_split(noise_samples,
                                                                                 noiseless_samples,
                                                                                 test_size=0.4, shuffle=True)
    
    x_test, x_valid, y_test, y_valid = train_test_split(x_test, y_test, test_size=0.5, shuffle=True)
    x_test_noise, x_valid_noise, y_test_noise, y_valid_noise = train_test_split(x_test_noise, y_test_noise,
                                                                                test_size=0.5, shuffle=True)
    
    x_train = np.concatenate((x_train, x_train_noise))
    y_train = np.concatenate((y_train, y_train_noise))

    x_valid = np.concatenate((x_valid, x_valid_noise))
    y_valid = np.concatenate((y_valid, y_valid_noise))

    x_test = np.concatenate((x_test, x_test_noise))
    y_test = np.concatenate((y_test, y_test_noise))

    del x_train_noise, y_train_noise, x_valid_noise, y_valid_noise, x_test_noise, y_test_noise
    del x_data, y_data, noise_samples, noiseless_samples

    for i in range(10):
        x_train, y_train = shuffle(x_train, y_train)
        x_valid, y_valid = shuffle(x_valid, y_valid)
        x_test, y_test = shuffle(x_test, y_test)
    
    dataset.append([[x_train, y_train], [x_valid, y_valid], [x_test, y_test]])
    print(x_train.shape, y_train.shape)
    print(x_valid.shape, y_valid.shape)
    print(x_test.shape, y_test.shape)
del data_adc_5_15
dataset = np.array(dataset, dtype=object)
```
